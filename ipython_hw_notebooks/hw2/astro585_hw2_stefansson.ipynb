#Question 2.

We are given the code below:

```julia
srand(42);
Nobs = 100;
z = zeros(Nobs);
sigma = 2. * ones(Nobs);
y = z + sigma .* randn(Nobs);
normal_pdf(z, y, sigma) = exp(-0.5*((y-z)/sigma)^2)/(sqrt(2*pi)*sigma);

function log_likelihood(y::Array, sigma::Array, z::Array)
   n = length(y);
   sum = zero(y[1]);
   for i in 1:n
    sum = sum + log(normal_pdf(y[i],z[i],sigma[i]));
   end;
   return sum;
end
```

```julia
log_likelihood(y,sigma,z)
```

##2a) Writing a benchmark function

```julia
function calc_log_likelihood(Nobs::Int, M::Int=1)
    #Evaluates the log_likelihood M times
    likelihoods = zeros(M)
    for i in 1:M
        srand(42);
        z = zeros(Nobs);
        sigma = 2. * ones(Nobs);
        y = z + sigma .* randn(Nobs);
        likelihoods[i] = log_likelihood(y,sigma,z);
    end
    return likelihoods
end

function calc_time_log_likelihood(Nobs::Int, M::Int=1)
    #Evaluates the time it tages to evaluate log_likelihood M times
    return (@elapsed calc_log_likelihood(Nobs,M))
end
```

```julia
function calc_log_likelihood_assert(Nobs::Int, M::Int=1)
    #Evaluates the log_likelihood M times
    likelihoods = zeros(M)
    for i in 1:M
        srand(42);
        z = zeros(Nobs);
        sigma = 2. * ones(Nobs);
        y = z + sigma.*randn(Nobs);
        #Adding assert statements
        @assert length(z) == length(y)
        @assert length(z) == length(sigma)
        #Beyond this point, calculations good
        likelihoods[i] = log_likelihood(y,sigma,z);
    end
    return likelihoods
end


function calc_time_log_likelihood_assert(Nobs::Int, M::Int=1)
    #Evaluates the time it tages to evaluate log_likelihood M times
    return (@elapsed calc_log_likelihood_assert(Nobs,M))
end
```

##2f) Rewriting - taking functions as arguments

```julia
function calc_log_likelihood_assert_func(func,Nobs::Int,M::Int=1)
    #Evaluates the log_likelihoo
    #The function f should be a function of y::Array, sigma::Array, z::Array,
    #i.e. we assume that it can be called in the following way
    # val = func(y::Array, sigma::Array, z::Array)
    # where:
    # - y       are the observed values
    # - sigma   are the uncertainties in the observations
    # - z       is the mean of the distribution
    
    likelihoods = zeros(M)
    for i in 1:M
        srand(42);
        z = zeros(Nobs);
        sigma = 2. * ones(Nobs);
        y = z + sigma.*randn(Nobs);
        #Adding assert statements
        @assert length(z) == length(y)
        @assert length(z) == length(sigma)
        #Beyond this point, calculations good
        likelihoods[i] = func(y,sigma,z);
    end
    return likelihoods
end


function calc_time_log_likelihood_assert_func(func,Nobs::Int, M::Int=1)
    #Evaluates the time it tages to evaluate log_likelihood M times
    return (@elapsed calc_log_likelihood_assert_func(func,Nobs,M))
end
```

```julia
time1 = calc_time_log_likelihood_assert_func(log_likelihood,10000,1001)
time2 = calc_time_log_likelihood_assert(10000,1001)
println("Calculating with 'function' approach' \t", time1)
println("Calculating with old approach' \t\t", time2)
```

We see that they take almost exactly the same time - so doing it in the more general manner (former one) is advisable - keeping code more modular, makes it easier to reuse!

##2g) Using our knowledge of the normal distribution
Now, lets avoid calling the exp function, by noticing that we are taking the log of an exponential, so they cancel each other out.

```julia
#Now lets plainly take the analytic log of normal_pdf
normal_pdf_no_exp(z, y, sigma) = -0.5*((y-z)/sigma)^2 - log(sqrt(2*pi)*sigma); 

function likelihood_no_exp(y::Array, sigma::Array, z::Array)
   n = length(y);
   sum = zero(y[1]);
   for i in 1:n
        sum = sum + normal_pdf_no_exp(y[i],z[i],sigma[i]);
   end;
   return sum;
end
```

And lets time it:

```julia
time1_no_exp = calc_time_log_likelihood_assert_func(likelihood_no_exp,10000,1001)
time2        = calc_time_log_likelihood_assert_func(log_likelihood,10000,1001)
res1_no_exp  = calc_log_likelihood_assert_func(likelihood_no_exp,10000,1001)
res2         = calc_log_likelihood_assert_func(log_likelihood,10000,1001)
println("Result 1: ", mean(res1_no_exp))
println("Result 2: ", mean(res2))
println("Time when calculating with 'likelihood_no_exp' \t", time1_no_exp)
println("Time when calculating with 'log_likelihood'' \t", time2)
```

We see that the former one is faster - so indeed, in the future it is good to keep in mind to not perform unneccesary computations that effectevily cancel out each other. A good method to improve performance

#Question 3)
Consider integrating the equations of motion for a test particle orbiting a star fixed at the origin.  In 2D, there are two 2nd order ODEs:  
$$\frac{d^2 r_x}{dt^2} = a_x = -GM r_x/r^3$$
$$\frac{d^2 r_y}{dt^2} = a_y = -GM r_y/r^3,$$
where $r = \sqrt(r_x^2+r_y^2)$ and $a_x$ and $a_y$ are the accelerations in the x and y directions.  
Numerically, it is generally easier to integrate 1st order ODEs, so we transform these into:
	\begin{eqnarray}
    d v_x/dt & = a_x & = -GM r_x/r^3  \\
	d v_y/dt & = a_y & = -GM r_y/r^3 \\
	d r_x/dt & = v_x & \\
	d r_y/dt & = v_y & \\
    \end{eqnarray}
Euler’s method is a simplistic algorithm for integrating a set of 1st order ODEs with state stored in a vector x:
$$
r(t_n+1) = r(t_n) + dt * \frac{dr}{dt}|_{t_n} \\
v(t_n+1) = v(t_n) + dt * \frac{dv}{dt}|_{t_n}
$$

##3a) Integrating using Euler's method
I ended up creating two other helper functions, to calculate the accelerations in both x and y directions

```julia
function accel_x(G, M, r_x, r_y)
    #Calculate the acceleration in the x direction
    val = -(G*M*r_x)/(sqrt(r_x^2+r_y^2))^3;
    return val;
end

    
function accel_y(G, M, r_x, r_y)
    #Calculate the acceleration in the y direction
    val = -(G*M*r_y)/(sqrt(r_x^2+r_y^2))^3;
    return val;
end


function integrate_euler!(state, dt, duration)
    #Performs euler integration
    # **state** is the initial state in the form (r_x, r_y, v_x, v_y)
    # **dt** is the timestep,
    # **duration** is the number of timesteps
    #RETURNS: A two dimensional array which contains the state values at each time step in the rows
    
    M = 1.0
    G = 1.0
    
    #Create an array which will contain the state values - in the rows - in the format:
    # (r_x, r_y, v_x, v_y) in each row. There is one timestep between each row-entry
    intval = zeros(duration,4)
            
    #Copy the initial state
    intval[1,:] = state
    
    for i in 2:duration
        #For v_x
        intval[i,3] = intval[i-1,3] + dt * accel_x(G, M, intval[i-1,1], intval[i-1,2])
        #For v_y
        intval[i,4] = intval[i-1,4] + dt * accel_y(G, M, intval[i-1,1], intval[i-1,2])
        #For r_x
        intval[i,1] = intval[i-1,1] + dt * intval[i,3]
        #For r_y
        intval[i,2] = intval[i-1,2] + dt * intval[i,4]
    end
    return intval
end
```

##3b) Integrating the system for 3 orbits, assuming G*M = 1, using 200 timesteps per orbit

```julia
intstate = [1,0,0,1]
points_per_orbit = 200
timestep = (2*pi)/points_per_orbit #200 points per orbit
num_orbits = 3
num_timesteps = num_orbits*points_per_orbit
orbits = integrate_euler!(intstate, timestep, num_timesteps)
```

The final state is very similar to the initial one, as expected, as we have traversed 3 complete orbits, returning at the initial point.

##3c) Plotting the trajectory


```julia
using PyPlot;
plot(orbits[:,1], orbits[:,2], color="red")
PyPlot.xlim([-1.5,1.5])
PyPlot.ylim([-1.5,1.5])
PyPlot.xlabel(r"$r_x$")
PyPlot.ylabel(r"$r_y$")
```

We see that the calculated trajectory is quite accurate (all three of them seem to coincide - which we expect when we don't account for drag/friction).

##3d) Leapfrog method
Write a new function integrate_leapfrog!(state,dt, duration) that is analogous to integrate_euler, but uses the “leapfrog” or modified Euler’s integration method:
$$
r(t_n+1/2) = r(t_n) + dt/2 * \frac{dr}{dt}|_{t_n} \\
v(t_n+1) = v(t_n) + dt * \frac{dv}{dt}|_{t_n+1/2} \\
r(t_n+1) = r(t_n+1/2) + dt/2 * \frac{dr}{dt}|_{t_n+1}
$$

Note that the equation for $r(t_{n+1})$ was wrong; I corrected it from 
$$r(t_n+1) = r(t_n+1/2) + dt/2 * \frac{dr}{dt}|_{t_n+1/2}$$ 
to
$$r(t_n+1) = r(t_n+1/2) + dt/2 * \frac{dr}{dt}|_{t_n+1}$$

```julia
function integrate_leapfrog!(state,dt,duration)
    #Performs leapfrog integration
    # **state** is the initial state in the form (r_x, r_y, v_x, v_y)
    # **dt** is the timestep,
    # **duration** is the number of timesteps
    
    M = 1.0
    G = 1.0
    
    #Create an array which will contain the state values - in the rows - in the format:
    # (r_x, r_y, v_x, v_y) in each row. There is one timestep between each row-entry
    intval = zeros(duration,4)
    
    #An array similar in the format: (r_x, r_y)
    # in each row. But row No. i in **halfsteps** corresponds to time t = i+0.5, i.e.
    # the necessary halfstep needed in the leap-frog method
    halfsteps = zeros(duration,2)
            
    #Copy the initial state
    intval[1,:] = state
    
    for i in 1:(duration-1)
        #For r_x at t_{i+0.5}
        halfsteps[i,1] = intval[i,1] + 0.5 * dt * intval[i,3]
        #For r_y at t_{i+0.5}
        halfsteps[i,2] = intval[i,2] + 0.5 * dt * intval[i,4]
        
        #For v_x at t_{i+1}
        intval[i+1,3] = intval[i,3] + dt * accel_x(G, M, halfsteps[i,1], halfsteps[i,2])
        #For v_y at t_{i+1}
        intval[i+1,4] = intval[i,4] + dt * accel_y(G, M, halfsteps[i,1], halfsteps[i,2])
        
        #For r_x at t_{i+1}
        intval[i+1,1] = halfsteps[i,1] + 0.5 * dt * intval[i+1,3]
        #For r_y at t_{i+1}
        intval[i+1,2] = halfsteps[i,2] + 0.5 * dt * intval[i+1,4]
    end
    return intval
end
```

##3e) Plotting, and checking the accuracy of the Leapfrod method
Using the same number of points as in the Euler method, we get

```julia
intstate = [1,0,0,1]
points_per_orbit = 200
timestep = (2*pi)/points_per_orbit #200 points per orbit
num_orbits = 3
num_timesteps = num_orbits*points_per_orbit

orbits_euler = integrate_euler!(intstate, timestep, num_timesteps)
orbits_frog = integrate_leapfrog!(intstate, timestep, num_timesteps)
println("Endstate Euler: ", orbits_euler[num_timesteps,:])
println("Endstate leapfrog: ", orbits_frog[num_timesteps,:])


using PyPlot;
plot(orbits_frog[:,1], orbits_frog[:,2], color="red")
PyPlot.xlim([-1.5,1.5])
PyPlot.ylim([-1.5,1.5])
PyPlot.xlabel(r"$r_x$")
PyPlot.ylabel(r"$r_y$")
```

We see that the endstate is very similar to the initial state: [1,0,0,1] as expected after 3 whole orbits. We see that the leapfrog method is marginally closer to 0 for r_y, and v_x.


As both methods perform very similarly for these number of points, lets try a few more points

```julia
intstate = [1,0,0,1]
points_per_orbit = 200
timestep = (2*pi)/points_per_orbit #200 points per orbit
num_orbits = 100
num_timesteps = num_orbits*points_per_orbit

orbits_euler = integrate_euler!(intstate, timestep, num_timesteps)
orbits_frog = integrate_leapfrog!(intstate, timestep, num_timesteps)
println("Endstate Euler: ", orbits_euler[num_timesteps,:])
println("Endstate leapfrog: ", orbits_frog[num_timesteps,:])
```

##3h) Log-Log plot of accuracy vs. integration duration

```julia
intstate = [1,0,0,1]
points_per_orbit = 200 
timestep = (2*pi)/points_per_orbit #200 points per orbit

array_num_orbits = [10,50,100,500,1000,5000]
orbit_samples = length(array_num_orbits)
array_num_timesteps = array_num_orbits*points_per_orbit

#Initialize data
accuracy = zeros(orbit_samples)
int_duration = zeros(orbit_samples)


function acc(a,b)
    #We will use the 4d geometric distance as a metric for the accuracy
    #a and b are two 4 dimensional vectors
    return sqrt((a[1]-b[1])^2 + (a[2]-b[2])^2 + (a[3]-b[3])^2 + (a[4]-b[4])^2);
end


for i in 1:orbit_samples
    println("Calculating orbit duration (i.e. number of orbits): ", array_num_timesteps[i])
    orbits_frog = integrate_leapfrog!(intstate, timestep, array_num_timesteps[i])

    init_state = orbits_frog[1,:]
    final_state = orbits_frog[array_num_timesteps[i],:]
    accuracy[i] = acc(init_state, final_state)
    int_duration[i] = (@elapsed integrate_leapfrog!(intstate, timestep, num_timesteps))
end

plot(log10(accuracy), log10(int_duration))
PyPlot.xlabel("Log(Integration duration in seconds)")
PyPlot.ylabel("Log(Accuracy - measured as geometric distance)")
```

Above we see a plot of the log(accuracy) as a function of log(integration time). I measured the accuracy as a geometric distance between the initial state and final state of the integrations.

One thing to note, that the curve constantly changed shape when I ran it again. From that we can conclude that as we increase the duration (the number of orbits), the system diverges more and more from the correct solution. We see the shift in points, due to the fact that the computing time varies. 

##3i) A few ideas how to accelerate the integrations
After reading a bit more about the **Leap Frog** method I saw that it can be formulated without using an intermediate array:
$$
x_{i+1} = x_i + v_i \Delta t + 0.5 a_i \Delta t^2
$$
$$
v_{i+1} = v_i + 0.5 (a_i + a_{i+1})\Delta t
$$


This should save computing time, as we are not reading/writing to a big intermediate array - but memory r&w is very time consuming (calculations are faster). 

We also might be able to accelerate the integrations by not saving all of the states, maybe only the initial and final ones.

It seems to me that making these computations parallel would be quite difficult, as the subsequent timestep depends on the one immediately before - it would therefore be interesting to check other algorithms Runge Kutta, etc.
