#Question 1 - Dense Matrix Multiply

##1a) Simple matrix vector multiplication

```julia
function multiply(b::Array{Float64,2},c::Array{Float64,1})
    return b*c
end
```

Lets try it out

```julia
srand(42)
A=rand(8,8);
x=rand(8);
C1=multiply(A,x);
print(C1)
```

##1b) Column multiply

```julia
function column_multiply(d::Array{Float64,2},e::Array{Float64,1})
    rows=length(e[:,1])
    columns=length(e[1,:])
    B=Array(Float64,rows,columns)
    for ii in 1:length(d[:,1])
        for jj in 1:length(d[ii,:])
            B[ii]=B[ii]+d[ii,jj]*e[jj]
        end
    end
    return B
end
```

Lets try it out

```julia
C2=column_multiply(A,x)
print(C2)
using Base.Test
print(@test_approx_eq_eps(C2,C1,0.001))
```

##1c) Row multiply

```julia
function row_multiply(d::Array{Float64,2},e::Array{Float64,1})
    rows=length(e[:,1])
    columns=length(e[1,:])
    B=Array(Float64,rows,columns)

    for ii in 1:length(d[1,:])
        for jj in 1:length(d[:,ii])
            B[ii]=B[ii]+d[ii,jj]*e[jj]
        end
    end
    return B
end
```

```julia
C3 = row_multiply(A,x)
print(C3)
println(@test_approx_eq_eps(C3,C2,0.001))
println(@test_approx_eq_eps(C3,C1,0.001))
```

##1d-e) Discussion about the above functions

We think that 1a would be the fastest, just because I would think the programmers that designed this would have optimized the commands for matrix multiplication. Moreover, we think for small matrices they will all have similar run times, but for large data sets, it will depend on how Julia stores the data for an array, either via rows or columns.

##1g-f) Benchmarking

####Small matrices

```julia
A=rand(8,8);
x=rand(8);
a2=@elapsed multiply(A,x);
b2=@elapsed row_multiply(A,x);
c2=@elapsed column_multiply(A,x);
println(a2)
println(b2)
println(c2)
```

####Bigger matrices

```julia
A=rand(1024,1024);
x=rand(1024);
a2=@elapsed multiply(A,x);
b2=@elapsed row_multiply(A,x);
c2=@elapsed column_multiply(A,x);
println(a2)
println(b2)
println(c2)
```

For the small matrix, the column multiplication is usually faster than the regular multiply function, by about a factor of 2 or so, (the regular multiply function is the slowest). For the large matrix, the row and column functions were comparable, but the regular matrix multiply function was the fastest. The regular multiply function is probably optimized better for larger matrices, and uses the cache lines better.

##2b) Creating a lookup table:

##2c) Plotting original function along with the approximation

We trained the table using 20 points, now lets try to see what happens if we use those 20 points to interpolate to more points:

```julia
using PyPlot;
```

```julia
yy=linspace(0.,2.,100);
#We use the version of the function "lookup" which takes
# a type of 'lookup_table_linear_type'-type as a first argument
approx=lookup(lookup_table,yy);
plot(yy,tanh(yy),color="blue",marker="*")
plot(yy,approx ,color="red", marker="+")
```

Lookup table works well, both curves look quite the same (easy case: the function is increasing monotonously).

##2d) Comparing performance

Lets see how the lookup-table performs with respect to plain function evaluations:

```julia
lookup_table = make_table_linear(x->tanh(x),0.,2.,1000);
n=100000;
x=linspace(0.,2.,n)
a=@elapsed lookup(lookup_table,x)
b=@elapsed tanh(x);
println(a)
println(b)
```

We see that the lookup table is slower for _tanh(x)_. Lets try a more complex function

```julia
function normal_pdf(x)
    #Returns the value of the normal_pdf at x
    return exp(-0.5*x.*x)/sqrt(2*pi)
end
```

```julia
lookup_table = make_table_linear(x->normal_pdf(x),-20.,20.,1000);
n=100000;
x=linspace(0.,2.,n)
a=@elapsed lookup(lookup_table,x)
b=@elapsed normal_pdf(x);
println(a)
println(b)
```

For a more complex function we see that the 'table-lookup'-method is relatively faster, (and its faster for the first run) but still slower than the plain function evaluation (for subsequent runs of the code) - perhaps our function isn't complex enough, or maybe this has something to do with the cache?

##2d') Kepler's equation
Lets plot the eccentric anomaly as a function of mean anomaly for an eccentricity of 0.25 :

```julia
include("HW5_Q2_ecc_anom.jl")
mn_am=linspace(-2pi,4pi,1000);
ecc=ecc_anom(mn_am,0.25);
plot(mn_am,ecc)
```

We think the lookup plot would have issues at the discontinuities, since it takes the derivative of the function. It would go crazy at those points.

##2e) Kepler's equation - using lookup tables
Lets see how the lookup table performs:

```julia
NN = 128;
#Lets make the lookup table
lookup_table = make_table_linear(x->ecc_anom(x,0.25),0.,2.*pi,NN);
NN=500;
mn_am  = linspace(0.,2.*pi,NN);
approx = lookup(lookup_table,mn_am)
plot(mn_am, approx,color="blue")
plot(mn_am,ecc_anom(mn_am,0.25),color="red")
```

The normally evaluated function is the red one, and the blue curve is the values we get from table-lookup. We see that the lookup table has difficulties at the discontinuities : the derivatives which are used in the linear interpolation are extremely large (approach infinity).

##2f) Exploring the first derivative

```julia
N = 128
lookup_table = make_table_linear(x->ecc_anom(x,0.25),x->decc_anom_dM(x,0.25),0.0,2.0*pi,N);
NN = 400;
mn_am  = linspace(0.,2.*pi,NN);
#calculating the approx:
approx = lookup(lookup_table,mn_am)
plot(mn_am, approx,color="blue")
plot(mn_am,ecc_anom(mn_am,0.25),color="red")
```

We see that the lookup table has difficulties near 4.2. Lets try to plot the derivative:

```julia
mn_am  = linspace(0.,2.*pi,NN);
derivVals = similar(mn_am)
for i in 1:NN
    derivVals[i] = decc_anom_dM(mn_am[i],0.25)
end
plot(mn_am,derivVals)
```

We see that the derivative has a singularity at around 4.2, explaining why the lookup-table has difficulties at around 4.2 as well.

##2g) Using higher order interpolation

Lets try using **make_table_quadratic** this time

```julia
NN = 128;
#Lets make the lookup table
lookup_table = make_table_quadratic(x->ecc_anom(x,0.25),0.,2.*pi,NN);
NN=400;
mn_am  = linspace(0.,2.*pi,NN);
approx = lookup(lookup_table,mn_am)
plot(mn_am, approx,color="blue")
plot(mn_am,ecc_anom(mn_am,0.25),color="red")
```

The lookup table still has problems at the derivative points.

##2h) Explaining problems
We should not use lookup tables (or interpolations) at or near discontinuities - i.e. lookup tables should be good for functions that take a long time to evaluate but are relatively slowly changing (have smooth derivatives that don't change too much).
